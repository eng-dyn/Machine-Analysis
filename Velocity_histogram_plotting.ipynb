```python
# Luff velocity histogram plotting

# Histogram settings
bins = 64
designL = 6   #design limit for luff velocity in m/min
ylim = max(round_to_nearest_log(Lv_array.size), round_to_nearest_log(LvO_array.size))
xlim = math.ceil(max(LTv_array.max(), LTvO_array.max()))

# Calculates the percentage and number of data points exceeding the design limit for both operating and non-operating conditions and prints the results
countL = np.sum(Lv_array > designL)
countLO = np.sum(LvO_array > designL)
print(f"Percentage over L: {countL/Lv_array.size*100:.3f}%", f"Percentage over L (Operating): {countLO/LvO_array.size*100:.3f}%")
print(f"Number over L: {countL}, Number over L (Operating): {countLO}")

# Channel appearance and titles
channel1 = [{"data": Lv_array, "label": "Non-Operating", "color": "#2B296B",   "alpha": 0.9}]
channel2 = [{"data": LvO_array, "label": "Operating", "color": "#0097D3",  "alpha": 0.6}]

# Title formatting
title_text = "Luff Speed"
title_fontsize = 16

# Axis formatting and title
xlabel_text = "Luff Speed (m/min R=56.8m)"
ylabel_text = "Frequency"
axis_label_fontsize = 13
tick_fontsize = 11

# Design limit bar settings
manual_bar = {
    "x": designL,          # x location of the bar
    "height": ylim,        # y value (height)
    "width": xlim/bins,    # bar width
    "color": "#d20000",
    "alpha": 0.6,
    "label": f"{designL}m/min"
}

# PLOTTING
fig, ax = plt.subplots(figsize=(10, 6), dpi=150)

# Plot histograms
#for ch in channel1:
ax.hist(channel1[0]["data"], bins=bins, range=(0, xlim), color=channel1[0]["color"], alpha=channel1[0]["alpha"], label=channel1[0]["label"], edgecolor="black", linewidth=0.6)
#for ch in channel2:
ax.hist(channel2[0]["data"], bins=bins, range=(0, xlim), color=channel2[0]["color"], alpha=channel2[0]["alpha"], label=channel2[0]["label"], edgecolor="black", linewidth=0.6)

# Plot manual bar
ax.bar(manual_bar["x"], manual_bar["height"], width=manual_bar["width"], color=manual_bar["color"], alpha=manual_bar["alpha"], label=manual_bar["label"], edgecolor="black", linewidth=0.6)

# Formatting
ax.set_title(title_text, fontsize=title_fontsize)
ax.set_xlabel(xlabel_text, fontsize=axis_label_fontsize)
ax.set_ylabel(ylabel_text, fontsize=axis_label_fontsize)
ax.set_yscale("log")
ax.set_ylim(1, ylim)
ax.set_xlim(0, xlim)
ax.tick_params(axis="both", labelsize=tick_fontsize)
ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=3,
    fontsize=10,
    frameon=False
)
plt.tight_layout()
plt.show()
```

```python
# Boom tip speed histogram plotting for collision cases FS and FT

# Histogram settings
bins = 64
design = 36   #design limit for long travel velocity

# Masking for FS and FT based on slew angle ranges
maskFS = (45 < slew_array) & (slew_array < 135) | (-45 > slew_array) & (slew_array > -135)
maskFSs = slew_array[maskFS]
maskFSLT = (LTv_array)[maskFS]
maskFSLTO = (LTvO_array)[maskFS]
maskFSS = (Sv_array)[maskFS]
maskFSSO = (SvO_array)[maskFS]
maskFT = (15 < slew_array) & (slew_array < 25) | (-15 > slew_array) & (slew_array > -25)
maskFTs = slew_array[maskFT]
maskFTLT = (LTv_array)[maskFT]
maskFTLTO = (LTvO_array)[maskFT]
maskFTS = (Sv_array)[maskFT]
maskFTSO = (SvO_array)[maskFT]

vFS = maskFSLT + maskFSS*np.cos(np.radians(maskFSs))
vFSO = maskFSLTO + maskFSSO*np.cos(np.radians(maskFSs))
vFT = maskFTLT + maskFTS*np.cos(np.radians(maskFTs))
vFTO = maskFTLTO + maskFTSO*np.cos(np.radians(maskFTs))

# Computing limits for plotting
ylimFS = max(round_to_nearest_log(vFS.size), round_to_nearest_log(vFSO.size))
xlimFS = math.ceil(max(vFS.max()+5, vFSO.max()+5))
ylimFT = max(round_to_nearest_log(vFT.size), round_to_nearest_log(vFTO.size))
xlimFT = math.ceil(max(vFT.max()+5, vFTO.max()+5))

# Calculates the percentage and number of data points exceeding the design limit for both operating and non-operating conditions and prints the results
countFS = np.sum(vFS > design)
countFSO = np.sum(vFSO > design)
countFT = np.sum(vFT > design)
countFTO = np.sum(vFTO > design)
print(f"Percentage over Design Limit (FS): {countFS/vFS.size*100:.3f}%", f"Percentage over Design Limit (Operating, FS): {countFSO/vFSO.size*100:.3f}%")
print(f"Number over Design Limit (FS): {countFS}, Number over Design Limit (Operating, FS): {countFSO}")
print(f"Percentage over Design Limit (FT): {countFT/vFT.size*100:.3f}%", f"Percentage over Design Limit (Operating, FT): {countFTO/vFTO.size*100:.3f}%")
print(f"Number over Design Limit (FT): {countFT}, Number over Design Limit (Operating, FT): {countFTO}")

# Channel appearance
channel1 = [{"data": vFS, "label": "Non-Operating (FS)", "color": "#2B296B",   "alpha": 0.9}]
channel2 = [{"data": vFSO, "label": "Operating (FS)", "color": "#0097D3",  "alpha": 0.7}]
channel3 = [{"data": vFT, "label": "Non-Operating (FT)", "color": "#2B296B",   "alpha": 0.9}]
channel4 = [{"data": vFTO, "label": "Operating (FT)", "color": "#0097D3",  "alpha": 0.7}]

# Title formatting
title_text1 = "Operating and Non-Operating Boom Tip Speed (FS)"
title_text2 = "Operating and Non-Operating Boom Tip Speed (FT)"
title_fontsize = 16

# Axis formatting and titles
xlabel_text = "Boom Tip Speed (m/min)"
ylabel_text = "Frequency"
axis_label_fontsize = 13
tick_fontsize = 11

# Design limits
manual_bar = {
    "x": design,           # x location of the bar
    "height": max(ylimFS, ylimFT),      # y value (height)
    "color": "#d20000",
    "alpha": 0.6,
    "label": f"Design Limit ({design} m/min)"
}

# PLOTTING
fig, ax = plt.subplots(figsize=(10, 6), dpi=150)

# Plot histograms
#for ch in channel1:
ax.hist(channel1[0]["data"], bins=bins, color=channel1[0]["color"], range=(0, xlimFS), alpha=channel1[0]["alpha"], label=channel1[0]["label"], edgecolor="black", linewidth=0.6)

#for ch in channel2:
ax.hist(channel2[0]["data"], bins=bins, color=channel2[0]["color"], range=(0, xlimFS), alpha=channel2[0]["alpha"], label=channel2[0]["label"], edgecolor="black", linewidth=0.6)

# Plot manual bar
ax.bar(manual_bar["x"], manual_bar["height"], width=xlimFS/bins, color=manual_bar["color"], alpha=manual_bar["alpha"], label=manual_bar["label"], edgecolor="black", linewidth=0.6)

# Formatting
ax.set_title(title_text1, fontsize=title_fontsize)
ax.set_xlabel(xlabel_text, fontsize=axis_label_fontsize)
ax.set_ylabel(ylabel_text, fontsize=axis_label_fontsize)
ax.set_yscale("log")
ax.set_ylim(1, ylimFS)
ax.set_xlim(0, xlimFS)
ax.tick_params(axis="both", labelsize=tick_fontsize)
ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=3,
    fontsize=10,
    frameon=False
)

plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(10, 6), dpi=150)

ax.hist(channel3[0]["data"], bins=bins, color=channel3[0]["color"], range=(0, xlimFT), alpha=channel3[0]["alpha"], label=channel3[0]["label"], edgecolor="black", linewidth=0.6)

#for ch in channel2:
ax.hist(channel4[0]["data"], bins=bins, color=channel4[0]["color"], range=(0, xlimFT), alpha=channel4[0]["alpha"], label=channel4[0]["label"], edgecolor="black", linewidth=0.6)

# Plot manual bar
ax.bar(manual_bar["x"], manual_bar["height"], width=xlimFT/bins, color=manual_bar["color"], alpha=manual_bar["alpha"], label=manual_bar["label"], edgecolor="black", linewidth=0.6)

# Formatting
ax.set_title(title_text2, fontsize=title_fontsize)
ax.set_xlabel(xlabel_text, fontsize=axis_label_fontsize)
ax.set_ylabel(ylabel_text, fontsize=axis_label_fontsize)
ax.set_yscale("log")
ax.set_ylim(1, ylimFT)
ax.set_xlim(0, xlimFT)
ax.tick_params(axis="both", labelsize=tick_fontsize)
ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=3,
    fontsize=10,
    frameon=False
)

plt.tight_layout()
plt.show()
```
